```python
# Filter for TAVG datatype
tavg_df = df[df['datatype'] == 'TAVG']
```

```python
# Group by 'state' and 'date', and take the mean of the 'value' column
grouped_tavg_df = tavg_df.groupby(['state', 'date'])['value'].mean().reset_index()
```

```python
grouped_tavg_df
```

```python
# Function to calculate monthly mean for each datatype

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

def calculate_monthly_mean(df):
    
     # Map month names to categorical data with custom sort order
    df['month'] = df['date'].dt.strftime('%B').astype(pd.CategoricalDtype(categories=month_order, ordered=True))
    
    # Group by state and month, and calculate the mean value for each month
    monthly_mean = df.groupby([df['state'], df['month']]).agg({'value': 'mean'}).reset_index()
    
    return monthly_mean
```

```python
prcp_monthly_mean = calculate_monthly_mean(grouped_prcp_df)
prcp_monthly_mean.head(30)
```

```python
prcp_monthly_mean = prcp_monthly_mean.rename(columns={'value': 'avg_prcp_per_month'})
prcp_monthly_mean.head(30)
```

```python
tavg_monthly_mean = calculate_monthly_mean(grouped_tavg_df)
tavg_monthly_mean = tavg_monthly_mean.rename(columns={'value': 'avg_tavg_per_month'})
tavg_monthly_mean.head(30)
```

```python
awnd_monthly_mean = calculate_monthly_mean(grouped_awnd_df)
awnd_monthly_mean = awnd_monthly_mean.rename(columns={'value': 'avg_awnd_per_month'})
awnd_monthly_mean.head(30)
```
